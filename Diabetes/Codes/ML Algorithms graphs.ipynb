```python
import matplotlib.pyplot as plt
import numpy as np

# Your data
before = [0.9374, 0.933, 0.933, 0.761, 0.935]
after = [0.94, 0.959, 0.9545, 0.841, 0.9573]
stack = [0.995, 0.99525, 0.995, 0.98143, 0.995]

# Attribute names
attributes = ['Accuracy','Precision','F1-Score','Kappa Index','Recall']

# Define the color for each sample
colors = ['royalblue','orangered','goldenrod']  # red for before, green for after, blue for stack

# Create an array with the positions of each bar along the x-axis
x = np.arange(len(attributes))

# Create the figure and the axes
fig, ax = plt.subplots()

# Plot each sample
ax.bar(x - 0.3, before, 0.3, color=colors[0], label='Before Attribute Selection')
ax.bar(x, after, 0.3, color=colors[1], label='After Attribute Selection')
ax.bar(x + 0.3, stack, 0.3, color=colors[2], label='After Stacking')

# Add a legend
ax.legend(loc = 'lower right')

# Replace the x-ticks with the attribute names
ax.set_xticks(x)
ax.set_xticklabels(attributes)

ax.set_title("Comparison of Results Obtained")
ax.set_xlabel("Evaluation Metrics")
ax.set_ylabel("Results")

plt.show()
```
